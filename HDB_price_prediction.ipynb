```python
import pandas as pd 
df  = pd.DataFrame([[2005, 400], [2010, 500],[2015, 700],[2020, 800],[2021, 1000]], columns=['year', 'cost of HDB room'])
#L, B = area
```

```python
df.head()
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
x = df['year'].values.reshape(-1,1)
y = df['cost of HDB room'].values.reshape(-1,1)
lr.fit(x,y)
```

```python
from matplotlib import pyplot as plt
plt.plot(x, lr.predict(x),color='k')
```

```python
lr.predict([[2025], [2030]])
```
